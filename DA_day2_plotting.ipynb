```python
#show the score comparison of two students in five different subjects
#subjects as x-axis marks as y-axis
s1=[10,20,30,40,50]
s2=[60,40,70,20,80]
sub=["maths","eng","soc","tel","hin"]
plt.plot(sub,s1,label="student1",color="cyan")
plt.plot(sub,s2,label="student2",color="green")
plt.title("marks of student")
plt.legend()
```

```python
s1=[10,20,30,40,50]
s2=[60,40,70,20,80]
sub=["maths","eng","soc","tel","hin"]
plt.subplot(2,1,1)
plt.plot(sub,s1,label="student1",color="green")
plt.title("marks of student")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,s2,label="student2",color="blue")
```

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
c=np.diff(A)
print(c)
D=np.diff(B)
print(D)
year=["2019-2020","2020-2021","2021-2022","2022-2023"]
plt.subplot(2,1,1)
plt.bar(year,c,label="profit A",color="k")
plt.legend()
plt.subplot(2,1,2)
plt.bar(year,D,label=" profit B",color="green")
plt.legend()
plt.show()
```
